```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
words = ["playing", "played", "plays", "happily", "studies"]
```

```python
stemmed_words = [stemmer.stem(word) for word in words]
print(stemmed_words)
```

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

words = ["running", "cars", "better", "children", "feet"]
```

```python
lemma_words = [lemmatizer.lemmatize(word) for word in words]
print(lemma_words)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words=['talking','charging','filling','saw','said']
```

```python
for i in words:
  s=stemmer.stem(i)
  l=lemmatizer.lemmatize(i)

  print(f"word: {i} |stem:{s}  |lemma:{l}\n")
```
